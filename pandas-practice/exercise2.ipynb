#How to handle missing data in Pandas

```python
salary = pd.read_csv("salary.csv", na_values= {'Salary':0.0})
```

```python
salary
```

Two Solution for handling missing value
1- Remove the row/rows
2- Filling out the missing values

```python
salary.dropna(axis=0, inplace=False, how="any")
```

```python
salary
```

```python
salary.dropna(axis=1, inplace=False, how="any")
```

```python
salary["Salary"].median()
```

```python
salary["Salary"].mode()
```

```python
salary["Salary"].mean()
```

```python
salary["Salary"].fillna(salary["Salary"].mean(), inplace=True)
```

```python
salary["Salary"].fillna(salary["Salary"].median(), inplace=False)
```

```python
salary["Salary"].fillna(80000, inplace=False)
```

```python
salary["Salary"].fillna(salary["Salary"].mean())
```

```python
salary
```

```python
salary["Salary"].fillna(method= "bfill", inplace=False)
```

```python
salary["Gender"].fillna("Female", inplace=True)
```

```python
salary
```

```python
salary.drop("Senior", axis=1, inplace=True)
```

```python
salary.dropna(axis=0, inplace=True, how="any")
```
